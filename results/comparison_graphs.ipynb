```python
import matplotlib.pyplot as plt
import matplotlib.collections as mcol
from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple
from matplotlib.lines import Line2D
import pandas as pd
import numpy as np
```

```python
#For Comparison of validation accuracy between the models
print("Comparison of Models for Architectural Evaluation, val accuracy\n")
df = pd.read_csv("val_acc_multimodal_inception_mobilenet.csv")

epochs = df["Epoch"].values
inception_val_acc = df["InceptionV3"].values
mobilenet_val_acc = df["MobileNetV2"].values

fig, ax = plt.subplots()

l1, l2 = ax.plot(epochs, inception_val_acc, epochs, mobilenet_val_acc)

ax.legend((l1, l2), ('InceptionV3', 'MobileNetV2'), loc='best', shadow=True)
ax.set_xlabel('Epochs')
ax.set_ylabel('Val Accuracy')
ax.set_title('Val Accuracy - InceptionV3, MobileNetV2')

plt.show()
```

```python
#For Comparison of validation loss between the three models
print("Comparison of Models for Architectural Evaluation, val loss\n")
df = pd.read_csv("val_loss_multimodal_inception_mobilenet.csv")

epochs = df["Epoch"].values
inception_val_acc = df["InceptionV3"].values
mobilenet_val_acc = df["MobileNetV2"].values

fig, ax = plt.subplots()

l1, l2 = ax.plot(epochs, inception_val_acc, epochs, mobilenet_val_acc)

ax.legend((l1, l2), ('InceptionV3', 'MobileNetV2'), loc='best', shadow=True)
ax.set_xlabel('Epochs')
ax.set_ylabel('Val Loss')
ax.set_title('Val Loss - InceptionV3, MobileNetV2')

plt.show()
```
