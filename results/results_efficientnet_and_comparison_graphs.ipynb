```python
import matplotlib.pyplot as plt
import matplotlib.collections as mcol
from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple
from matplotlib.lines import Line2D
import pandas as pd
import numpy as np
```

```python
def plot_df(filename, title):
    df = pd.read_csv(filename) 
    epochs = df["Epoch"].values
    train_acc = df["train Accuracy"].values
    val_acc = df["val Accuracy"].values
    train_loss = df["train Loss"].values
    val_loss = df["val Loss"].values

    xmax = epochs[np.argmax(val_acc)]
    ymax = val_acc.max()
    text= "x={:.3f}, y={:.3f}".format(xmax, ymax)

    fig, ax = plt.subplots()

    l1, l2 = ax.plot(epochs, train_acc, epochs, val_acc)

    ax.legend((l1, l2), ('Train Accuracy', 'Val Accuracy'), loc='lower right', shadow=True)
    ax.set_xlabel('Epochs')
    ax.set_ylabel('Accuracy')
    ax.set_title(title)

    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="left", va="bottom") #Changed from right and top to left and bottom
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.975), **kw)
    plt.show()
    print("")
```

```python
def get_val_acc(filename):
    df = pd.read_csv(filename)
    epochs = df["Epoch"].values
    val_acc = df["val Accuracy"].values
    
    return epochs, val_acc
```

```python
print("EfficientNet - Train and Val Accuracies.")
print("Batch size = 32, lr=0.001")
plot_df("48epochs_lr_0_001_efficient_bs16.csv",'Accuracy vs Epochs - EfficientNetB7')
```

```python
#For Comparison of validation accuracy between the three models
print("Bert-Image multimodal learning, Comparison of Models, val accuracy\n")
df = pd.read_csv("val_acc_multimodal_inception_efficientnet_mobilenet.csv")

epochs = df["Epoch"].values
inception_val_acc = df["InceptionV3"].values
efficientnet_val_acc = df["EfficientNetB7"].values
mobilenet_val_acc = df["MobileNetV2"]

fig, ax = plt.subplots()

l1, l2, l3 = ax.plot(epochs, inception_val_acc, epochs, efficientnet_val_acc, epochs, mobilenet_val_acc)

ax.legend((l1, l2, l3), ('InceptionV3', 'EfficientNetB7', 'MobileNetV2'), loc='best', shadow=True)
ax.set_xlabel('Epochs')
ax.set_ylabel('Val Accuracy')
ax.set_title('Val Accuracy - InceptionV3, EfficientNetB7, MobileNetV2')

plt.show()
```

```python
#For Comparison of validation loss between the three models
print("Bert-Image multimodal learning, Comparison of Models, val loss\n")
df = pd.read_csv("val_loss_multimodal_inception_efficientnet_mobilenet.csv")

epochs = df["Epoch"].values
inception_val_acc = df["InceptionV3"].values
efficientnet_val_acc = df["EfficientNetB7"].values
mobilenet_val_acc = df["MobileNetV2"]

fig, ax = plt.subplots()

l1, l2, l3 = ax.plot(epochs, inception_val_acc, epochs, efficientnet_val_acc, epochs, mobilenet_val_acc)

ax.legend((l1, l2, l3), ('InceptionV3', 'EfficientNetB7', 'MobileNetV2'), loc='best', shadow=True)
ax.set_xlabel('Epochs')
ax.set_ylabel('Val Loss')
ax.set_title('Val Loss - InceptionV3, EfficientNetB7, MobileNetV2')

plt.show()
```
